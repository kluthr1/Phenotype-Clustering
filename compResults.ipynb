```python
df = pd.read_csv('./finalResults.txt', sep="\t")
df
```

```python
def different1(val1, val2):
    val1 = float(val1)
    val2 = float(val2)
    if(val1 < 4 and val2 > 6):
        return False
    elif(val1 > 6 and val2 <4):
        return False
    return True

def different2(val1, val2):
    val1 = float(val1)
    val2 = float(val2)
    return abs(val1 - val2) < 3

```

```python
df['different1'] = df.apply(lambda x: different1(x['User Entry'], x['Alogrithm Entry']), axis = 1)
```

```python
df['different2'] = df.apply(lambda x: different2(x['User Entry'], x['Alogrithm Entry']), axis = 1)
```
